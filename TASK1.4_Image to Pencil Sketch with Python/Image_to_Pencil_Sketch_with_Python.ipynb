1.4:Image to Pencil Sketch with Python

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/drive/MyDrive/LGM/house.jpeg")
cv2_imshow(image)
cv2.waitKey(0)
```

```python
converted_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(converted_image)
cv2.waitKey(0)
```

```python
inverted_image = 255 - converted_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
inverted_blurred = 255 - blurred
```

```python
pencil_sketch_image = cv2.divide(converted_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch_image)
cv2.waitKey(0)
```
